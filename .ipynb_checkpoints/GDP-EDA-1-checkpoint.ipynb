## How many Unique Countries are there in the dataset?

```python
res=df["Country Name"].nunique()
print(f"There are {res} unique countries in the given dataset.")
```

## How many years of data we are having for India?

```python
df_ind=df[df["Country Name"]=="India"]
```

```python
df_ind.head()
```

```python
df_ind.nunique()
```

```python
res=len(df_ind)
res
```

```python
print(f"We are having {res} years of GDP Data for India")
```

## What is the World GDP in 1969?

```python
df_1969=df[df["Year"]==1969]
```

```python
df_1969
```

```python
res=df_1969["Value"].sum()
print(f"The World GDP in the year 1969 was {res}")
```

## What is the Percent of growth of GDP of India in 1994 as compared to 1993?

```python
GDP_1993=df_ind[df_ind["Year"]==1993]["Value"].values[0]
```

```python
GDP_1993
```

```python
GDP_1994=df_ind[df_ind["Year"]==1994]["Value"].values[0]
```

```python
GDP_1994
```

```python
diff=((GDP_1994-GDP_1993)/GDP_1993)*100
```

```python
print(f"There is {diff:.2f}% of Growth of GDP of India in 1994 as Compared to 1993")
```

 ## Find GDP Growth of India from 1960 - 2016

```python
df_ind=df_ind.copy()
df_ind["GDP_Growth"]=df_ind["Value"].pct_change()*100
```

```python
df_ind
```

```python
df_ind.plot(kind="line",x="Year",y="GDP_Growth")
```
